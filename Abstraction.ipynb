**Computing Employee Bonuses**

*   To make things entertaining, we make up a story 
*   You work for Microsoft
*   Microsoft sells ads, and is in the cloud business
*   If the company makes at least 10 Billion, the employees get a bonus


Code-wise, we are interested in this condition:

```
revenue_cloud + revenue_ads >= 10**10
```

For debugging, we will also print:

```
revenue_cloud + revenue_ads
```

````python
# A function to decide if employees get a bonus
# 10**10 is Ten Billion
def DoEmployeesGetABonus(revenue_cloud, revenue_ads):
  print("The sum of Cloud and Ads Revenue is:", revenue_cloud + revenue_ads)
  if (revenue_cloud + revenue_ads >= 10**10): 
    print("Everyone gets a Bonus!")
  else:
    print("No Bonus :(")
````

It is useful to test **boundary conditions**.
Therefore we consider three cases

*   9999999990 + 10 (Ans: **bonus**!)
*   9999999990 + 9.999999  (Ans:**no bonus**!)
*   9999999990 + 9.9999999  (Ans:**no  bonus**!)

*   It is time to discuss why a float is not like a real number
*   A [float](https://en.wikipedia.org/wiki/Floating-point_arithmetic) is a representation that trades precision for storage and speed. 
*   A float is an **approximate** representation; **many nearby numbers share the same representation.**
*   By doing this, the computer can store infinitely many numbers in finite, bounded space
*   Every number in between two green lines is represented by the line (number) nearest to it. Only numbers on the green lines are represented perfectly

* Notice that numbers near zero are represented more precisely, and numbers away from zero, less precisely

**Floating point numbers**

**Case 3: Integers in Numpy**

*   Numpy is a Python **library** that helps with large mathematical operations E.g. Matrix multiplication 
*   Turns out that simple Python isn't efficient for this
*   A library is a collection of helpful functions
*   Let us see what happens when we add numbers in Numpy



````python
import numpy as np

revenue_cloud = np.int_(9999999990)

revenue_ads = np.int_(10)

DoEmployeesGetABonus(revenue_ads, revenue_cloud)
````

* This looks good!
* Now let us specify the amount of space used by numpy to store integers
* Let us start with 32 bits

````python

revenue_cloud = np.int32(9999999990)

revenue_ads = np.int32(10)

DoEmployeesGetABonus(revenue_ads, revenue_cloud)



````

* Yikes, something went wrong! Just look at the sum!!?
* Let us increase the space

````python
revenue_cloud = np.int64(9999999990)

revenue_ads = np.int64(10)

DoEmployeesGetABonus(revenue_cloud, revenue_ads)
````

* This looks fine.
* Let us reduce the space to 16 bits

````python
revenue_cloud = np.int16(9999999990)

revenue_ads = np.int16(10)

DoEmployeesGetABonus(revenue_ads, revenue_cloud)
````

* It doesn't work. 
* But look at the debug. The sum is **negative!!**
* Turns out, we can **tell** numpy that the number should be positive or **unsigned** by making the 'int' a 'uint'

````python
revenue_cloud = np.uint16(9999999990)

revenue_ads = np.uint16(10)

DoEmployeesGetABonus(revenue_ads, revenue_cloud)
````

* The answer is wrong, but the number is positive.
* What is going on?
* Let us explain. 
* Numpy is implemented in C/C++. 
* It uses C's model of numbers (And you encounter this even though you are using Python!!)
* In C, integers are represented in finite amount of space. 
* Of course, in a finite amount of space, you cannot represent infinitely many numbers
* It turns out that the biggest positive number that you can represent in 16 bits is 32767, in 32 bits 2,147,483,647, and in 64 bits, 9,223,372,036,854,775,807. 
* When you add numbers, this very strange phenomenon called **overflow** happens







**Overflow**

* Overflow is when the number wraps around. So if you have a integer 32768, one more than the maximum representable in 16 bits, it wraps around to the smallest **negative** number -32768. (Interestingly the smallest negative number is larger in magnitude than the biggest positive number; Why?)
* On the other hand, if we make the number unsigned, the biggest positive number you can represent is 65535. 65536 becomes the same as zero
* You can think of the unsigned case like a Car's Odometer

Odometer_rollover.jpg



````python
print(np.int16(32768))
print(np.int16(65536))
````

# Bits and Gates

* We mentioned **bits** earlier; let us discuss exactly what they are
* Computers are built using transistors
* Transistors are two states devices
* It is therefore useful to represent numbers in a two digit system: 0,1
* This is called the **bi**nary system, and a bit is a **bi**nary digi**t**
* Every bit is implemented by a device called a [flip-flop](https://en.wikipedia.org/wiki/Flip-flop_(electronics) that uses a couple of transistors
* A binary number is just a string of bits; just as a decimal number is a string of digits
* Let us examine some decimal numbers in binary
* Ignore the preceding '0b'; that is python's way of saying the number is binary  


````python
def Bin(x):
  print(x,"becomes",bin(x), "in binary") 

Bin(0)
Bin(1)
Bin(2)
Bin(3)
Bin(4)
Bin(5)
Bin(8)
Bin(16)
Bin(65536)
Bin(65535)
````

** Rules for Carry**



*   There are three inputs: the bit D1 from the first number, bit D2 from the second number and the carry in. 
*   There are two outputs. The carry out, and the result. 
*   Let us work out the logic. 
* The **result** is '1' if there are an odd number of ones in the input. Else it is zero.
* The **carry out** is 1 if there are two or more ones in the input; zero otherwise




* Logic in computers are implemented using **Gates**

**Gates in Python**
* We are going to use Python to explain the operation of Gates
* This is funny: We implement gates in python, and python runs on a computer that is implemented using gates.
* But it is very interesting (and common for a computer scientist) to use a computer to think, and that is what we are going to do
* In Python, a bit can be represented as a **bool** type, and you have the operators 'or', 'and' and 'not'



````python
a = True
b = False

print(not a)
print (a and b)
print (a or b)
````

We can use these gates to implement other operators
Let us implement the NAND and XOR operators


````python
def nand(a,b):
  return not(a and b)

print(nand(True,True))
print(nand(True, False))
print(nand(False,True))
print(nand(False, False))


def xor(a,b):
  return nand(a,b) and (a or b)

print(xor(True,True))
print(xor(True, False))
print(xor(False,True))
print(xor(False, False))  

````

* Interestingly, the nand gate is a **universal gate**, it can implement all other gates
* Let us implement a couple of gates


````python
def not_using_nand(a):
  return nand(a,a)

print(not_using_nand(True))
print(not_using_nand(False))

def or_using_nand(a,b):
  return nand(nand(a,a),nand(b,b))

print(nand(True, True))
print(nand(True, False))
print(nand(False, True))
print(nand(False, False))
````

* We can also implement the carry logic using these gates
* First, let us implement the logic for one position



````python
def result_and_carry(bit_a, bit_b, carry_in):
  # carry out is true if any two of the three inputs are True
  carry_out_bit = (bit_a and bit_b) or (bit_a and carry_in) or (bit_b and carry_in)
  # result is true when one or three of the three inputs are True
  result_bit = (bit_a and not bit_b and not carry_in) or 
  (not bit_a and bit_b and not carry_in) or 
  (not bit_a and not bit_b and carry_in) or 
  (bit_a and bit_b and carry_in)
  return result_bit, carry_out_bit

print(result_and_carry(True, True, True))
print(result_and_carry(True, False, True))
print(result_and_carry(False, False, True))

````

````python
def binary_addition(a,b):
  # Why the reverse?
  a.reverse()
  b.reverse()
  result = [False] *len(a)
  carry_in = False
  for i in range(len(a)):
    result_bit, carry_out_bit = result_and_carry(a[i],b[i], carry_in)
    result[i] = result_bit
    carry_in = carry_out_bit
  result.reverse()  
  return result

a = [False, True, True, True]
b = [False, False, False, True]
print(binary_addition(a,b))


a = [False, True, False, True]
b = [False, False, True, False]
print(binary_addition(a,b))


# Overflow!
a = [True, True, True, False]
b = [False, False, True, False]
print(binary_addition(a,b))

````
